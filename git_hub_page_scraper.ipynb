# FINAL CODE

```python
def get_repo_info(h2_tag , star_tag):
    a_tags = h2_tag.find_all('a')
    username = a_tags[0].text.strip() 
    repo_name = a_tags[1].text.strip()
    repo_url = base_url+a_tags[1]['href']
    stars = parse_star_count(star_tag.text.strip())
    return username , repo_name,stars , repo_url

def get_topic_repos(topic_page_url):
    # page url
    response = requests.get(topic_page_url)
    # check response
    if(response.status_code !=200):
        raise Exception('Failed to load page {}'.format(topic_page_url))
    #parse using beautiful soup
    topic_doc = BeautifulSoup(response.text,'html.parser')
    # get the tag containing the information
    repo_tags = topic_doc.find_all('h3',{'class':'f3 color-text-secondary text-normal lh-condensed'})
    star_tags = topic_doc.find_all('a',{'class':'social-count float-none'})
    # dictionary to append the information
    topic_repo_dict = {
    'username':[],
    'repo_name':[],
    'stars':[],
    'repo_url':[]
    }
    # get all the info appended on the dictionary
    for i in range(len(repo_tags)):
        repo_info = get_repo_info(repo_tags[i] , star_tags[i])
        topic_repo_dict['username'].append(repo_info[0])
        topic_repo_dict['repo_name'].append(repo_info[1])
        topic_repo_dict['stars'].append(repo_info[2])
        topic_repo_dict['repo_url'].append(repo_info[3])
    return pd.DataFrame(topic_repo_dict)
```
